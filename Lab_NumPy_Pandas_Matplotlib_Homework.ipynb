## **Phần 1: NumPy cơ bản**

### **Bài tập 1: Tạo mảng và thao tác cơ bản**

1 Tạo một mảng NumPy với các giá trị từ 1 đến 20.\
2 Tìm tổng, giá trị lớn nhất, nhỏ nhất và trung bình của mảng.\
3 Tạo một mảng 2D (3x5) chứa các số ngẫu nhiên từ 0 đến 100.\
4 Lấy hàng thứ 2 và cột thứ 3 của mảng 2D.

```python
import numpy as np

# 1. Tạo một mảng NumPy với các giá trị từ 1 đến 20
array_1 = np.arange(1, 21)
print("Mảng từ 1 đến 20:", array_1)

# 2. Tìm tổng, giá trị lớn nhất, nhỏ nhất và trung bình của mảng
array_sum = np.sum(array_1)
array_max = np.max(array_1)
array_min = np.min(array_1)
array_mean = np.mean(array_1)

print("Tổng:", array_sum)
print("Giá trị lớn nhất:", array_max)
print("Giá trị nhỏ nhất:", array_min)
print("Trung bình:", array_mean)

# 3. Tạo một mảng 2D (3x5) chứa các số ngẫu nhiên từ 0 đến 100
array_2d = np.random.randint(0, 100, size=(3, 5))
print("Mảng 2D:", array_2d)

# 4. Lấy hàng thứ 2 và cột thứ 3 của mảng 2D
row_2 = array_2d[1]  # Hàng thứ 2 (index 1)
col_3 = array_2d[:, 2]  # Cột thứ 3 (index 2)

print("Hàng thứ 2:", row_2)
print("Cột thứ 3:", col_3)
```

### **Bài tập 2: Các thao tác nâng cao**

1 Tạo một mảng NumPy chứa 20 giá trị ngẫu nhiên từ 0 đến 1.\
2 Chuẩn hóa mảng này (đưa các giá trị về khoảng [0, 1]).\
3 Tính tích vô hướng (dot product) của hai mảng 1D: [1, 2, 3] và [4, 5,6].\
4 Tạo một ma trận 5x5 và tính định thức (determinant) và nghịch đảo của ma trận.

```python

import numpy as np

# 1. Tạo một mảng NumPy chứa 20 giá trị ngẫu nhiên từ 0 đến 1
random_array = np.random.rand(20)
print("Mảng ngẫu nhiên từ 0 đến 1:", random_array)

# 2. Chuẩn hóa mảng này (đưa các giá trị về khoảng [0, 1])
min_val = np.min(random_array)
max_val = np.max(random_array)
normalized_array = (random_array - min_val) / (max_val - min_val)
print("Mảng sau khi chuẩn hóa:", normalized_array)

# 3. Tính tích vô hướng (dot product) của hai mảng 1D
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
dot_product = np.dot(a, b)
print("Tích vô hướng của a và b:", dot_product)

# 4. Tạo một ma trận 5x5 và tính định thức, nghịch đảo
matrix = np.random.randint(1, 10, (5, 5))
print("Ma trận:", matrix)

# Tính định thức
determinant = np.linalg.det(matrix)
print("Định thức của ma trận:", determinant)

# Kiểm tra tính khả nghịch và tính nghịch đảo nếu có
if determinant != 0:
    inverse_matrix = np.linalg.inv(matrix)
    print("Ma trận nghịch đảo:", inverse_matrix)
else:
    print("Ma trận không khả nghịch (định thức = 0).")

```

## **Phần 2: Pandas cơ bản**

### **Bài tập 3: Làm quen với DataFrame**
1 Tạo một DataFrame chứa thông tin sau:
| Name    | Age | Score |
|---------|-----|-------|
| Alice   | 23  | 85    |
| Bob     | 25  | 90    |
| Charlie | 22  | 78    |
| David   | 24  | 92    |
| Eva     | 21  | 88    |

2 Tính giá trị trung bình của cột "Score".\
3 Lọc các hàng có "Score" lớn hơn 85.

```python
import pandas as pd

# 1. Tạo DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Age": [23, 25, 22, 24, 21],
    "Score": [85, 90, 78, 92, 88]
}
df = pd.DataFrame(data)
print("DataFrame:")
print(df)

# 2. Tính giá trị trung bình của cột "Score"
mean_score = df["Score"].mean()
print("Giá trị trung bình của cột 'Score':", mean_score)

# 3. Lọc các hàng có "Score" lớn hơn 85
filtered_df = df[df["Score"] > 85]
print("Các hàng có Score > 85:")
print(filtered_df)
```

### **Bài tập 4: Đọc và phân tích dữ liệu từ file**

1 Tải file Iris.csv từ Kaggle Iris Dataset.\
2 Đọc dữ liệu từ file CSV vào DataFrame.\
3 Hiển thị thông tin cơ bản (tổng quan, kiểu dữ liệu, số lượng null).\
4 Tính trung bình, lớn nhất, nhỏ nhất của cột sepal_length.

```python

import pandas as pd

# 1. Đọc dữ liệu từ file CSV vào DataFrame
iris_df = pd.read_csv('Iris.csv')  # Đảm bảo rằng đường dẫn đến file là đúng
print("Thông tin tổng quan về dữ liệu:")
print(iris_df.info())  # Hiển thị thông tin cơ bản về DataFrame
print("\nMô tả dữ liệu:")
print(iris_df.describe())  # Mô tả thống kê của các biến số

# 4. Tính toán cơ bản
mean_sepal_length = iris_df['SepalLengthCm'].mean()  # Tính trung bình
max_sepal_length = iris_df['SepalLengthCm'].max()    # Giá trị lớn nhất
min_sepal_length = iris_df['SepalLengthCm'].min()    # Giá trị nhỏ nhất

print("\nTrung bình SepalLengthCm:", mean_sepal_length)
print("Giá trị lớn nhất SepalLengthCm:", max_sepal_length)
print("Giá trị nhỏ nhất SepalLengthCm:", min_sepal_length)
```

## **Phần 3: Làm sạch dữ liệu**

### **Bài tập 5: Xử lý dữ liệu thiếu**
1 Tạo một DataFrame chứa các giá trị sau:

| Name    | Age | City     | Salary |
|---------|-----|----------|--------|
| Alice   | 23  | New York | 60000  |
| Bob     | NaN | Boston   | 52000  |
| Charlie | 25  | NaN      | NaN    |
| David   | 24  | Chicago  | 58000  |
| Eva     | 22  | Boston   | NaN    |

2 Điền giá trị thiếu trong cột Age bằng giá trị trung bình.\
3 Xóa các hàng có nhiều hơn 1 giá trị thiếu.\
4 Điền giá trị thiếu trong cột Salary bằng 50000.

```python
import pandas as pd
import numpy as np

# Tạo DataFrame chứa dữ liệu thiếu
data_with_missing = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [23, np.nan, 25, 24, 22],
    'City': ['New York', 'Boston', np.nan, 'Chicago', 'Boston'],
    'Salary': [60000, 52000, np.nan, 58000, np.nan]
}

df_missing = pd.DataFrame(data_with_missing)
print("Dữ liệu ban đầu:")
print(df_missing)

# 1. Điền giá trị thiếu trong cột Age bằng giá trị trung bình
mean_age = df_missing['Age'].mean()  # Tính giá trị trung bình của cột Age
df_missing['Age'].fillna(mean_age, inplace=True)  # Điền giá trị thiếu bằng giá trị trung bình
print("\nDataFrame sau khi điền giá trị thiếu cho cột Age:")
print(df_missing)

# 2. Xóa các hàng có nhiều hơn 1 giá trị thiếu
df_missing = df_missing.dropna(thresh=len(df_missing.columns) - 1)  # Giữ lại các hàng mà có ít nhất (tổng số cột - 1) giá trị không thiếu
print("\nDataFrame sau khi xóa các hàng có nhiều hơn 1 giá trị thiếu:")
print(df_missing)

# 3. Điền giá trị thiếu trong cột Salary bằng 50000
df_missing['Salary'].fillna(50000, inplace=True)  # Điền giá trị thiếu trong cột Salary bằng 50000
print("\nDataFrame sau khi điền giá trị thiếu cho cột Salary:")
print(df_missing)
```

## **Phần 4: Trực quan hóa dữ liệu với Matplotlib**

### **Bài tập 6: Biểu đồ cơ bản**

1 Tạo một biểu đồ đường biểu diễn hàm số y = x ** 2 trên khoảng [-10,10] \
2 Vẽ biểu đồ cột thể hiện điểm số (Score) của các sinh viên từ Bài tập 3.\
3 Tạo một biểu đồ tròn (pie chart) thể hiện phần trăm mỗi loại hoa trong tập dữ liệu Iris.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 1. Biểu đồ đường hàm số y = x^2
x = np.linspace(-10, 10, 100)  # Tạo 100 điểm x từ -10 đến 10
y = x ** 2  # Tính giá trị y = x^2

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='y = x^2', color='green')  # Vẽ biểu đồ đường
plt.title('Biểu đồ đường hàm số y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5, ls='--')  # Vẽ trục x
plt.axvline(0, color='black',linewidth=0.5, ls='--')  # Vẽ trục y
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()

# 2. Biểu đồ cột điểm số
# Giả lập dữ liệu điểm số của sinh viên (điểm số giả định)
students = ['Alice', 'Bob', 'Charlie', 'David', 'Eva']
scores = [85, 90, 78, 92, 88]

plt.figure(figsize=(10, 5))
plt.bar(students, scores, color='green')  # Vẽ biểu đồ cột
plt.title('Biểu đồ cột điểm số của sinh viên')
plt.xlabel('Tên sinh viên')
plt.ylabel('Điểm số')
plt.ylim(0, 100)  # Đặt giới hạn trục y từ 0 đến 100
plt.grid(axis='y', linestyle='--', linewidth=0.5)  # Vẽ lưới cho trục y
plt.show()

# 3. Biểu đồ tròn thể hiện phần trăm mỗi loại hoa trong tập dữ liệu Iris
# Tạo dữ liệu mẫu cho tập dữ liệu Iris
iris_data = {
    'Species': ['Setosa', 'Versicolor', 'Virginica'],
    'Count': [50, 50, 50]  # Giả định có 50 mẫu cho mỗi loại hoa
}

iris_df = pd.DataFrame(iris_data)

plt.figure(figsize=(7, 7))
plt.pie(iris_df['Count'], labels=iris_df['Species'], autopct='%1.1f%%', colors=['lightblue', 'lightcoral', 'lightgreen'])
plt.title('Phần trăm mỗi loại hoa trong tập dữ liệu Iris')
plt.axis('equal')  # Đảm bảo biểu đồ tròn là hình tròn
plt.show()
```

### **Bài tập 7: Biểu đồ nâng cao**

1 Vẽ biểu đồ phân tán (scatter plot) giữa sepal_length và sepal_width của tập dữ liệu Iris. Dùng màu sắc để phân biệt các loại hoa (species).\
2 Thêm tiêu đề, nhãn trục và chú thích cho biểu đồ.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file CSV (đảm bảo rằng đường dẫn đến file là đúng)
iris_df = pd.read_csv('Iris.csv')

# 1. Biểu đồ phân tán với màu sắc theo loại hoa
plt.figure(figsize=(10, 6))

# Vẽ biểu đồ phân tán với colormap mới
scatter = plt.scatter(iris_df['SepalLengthCm'], iris_df['SepalWidthCm'],
                      c=iris_df['Species'].astype('category').cat.codes,  # Chuyển đổi thành mã số cho màu sắc
                      cmap='plasma',  # Thay đổi colormap từ 'viridis' sang 'plasma'
                      edgecolor='k')  # Đường viền màu đen cho điểm

# Thêm tiêu đề và nhãn trục
plt.title('Biểu đồ phân tán giữa Sepal Length và Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

# Thêm chú thích
legend1 = plt.legend(*scatter.legend_elements(), title="Loại hoa")
plt.axis('equal')  
plt.grid(True) 
plt.show()
```
